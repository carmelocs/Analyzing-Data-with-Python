```python
N = int(input())
X = list(map(int, input().split()))
print(N, X)
```

__Pure Python__

```python
mean = sum(X)/N

X.sort()

median = (X[int(N/2-1)]+X[int(N/2)])/2 if N%2==0 else X[int(N//2)]

count = [X.count(cnt) for cnt in X]
data = {number:cnt for (number, cnt) in zip(X, count)}
mode = [k for k, v in data.items() if v==max(data.values())][0]

print(f'{mean:.1f}\n{median}\n{mode}')
```

__Pure Python__

```python
def median(X):
    return (X[int(len(X)/2-1)]+X[int(len(X)/2)])/2 if len(X)%2==0 else X[int(len(X)//2)]

def quartiles(arr):
    arr.sort()
    q2 = median(arr)
    if len(arr)%2==1:
        lower_arr = arr[:len(arr)//2]
        q1 = median(lower_arr)
        upper_arr = arr[(len(arr)+1)//2:]
        q3 = median(upper_arr)
    else:
        lower_arr = arr[:len(arr)//2]
        q1 = median(lower_arr)
        upper_arr = arr[len(arr)//2:]
        q3 = median(upper_arr)

    return q1, q2, q3
```

```python
q1, q2, q3 = quartiles(data)
```

__Solution__

```python
n = int(input().strip())

X = list(map(float, input().rstrip().split()))

Y = list(map(float, input().rstrip().split()))

print(n, X, Y)
```

__Python Implementation__

```python
from math import sqrt

X_mean = sum(X)/n
Y_mean = sum(Y)/n
X_std = sqrt(sum([(x-X_mean)**2 for x in X])/n)
Y_std = sqrt(sum([(y-Y_mean)**2 for y in Y])/n)
```

```python
X1 = [x-X_mean for x in X]
Y1 = [y-Y_mean for y in Y]
XY = sum([x*y for x, y in zip(X1, Y1)])
pearson_coef = XY/(n*X_std*Y_std)
print(round(pearson_coef, 3))
```

__Numpy Implementation__

```python
import numpy as np

pearson_coef = sum((np.array(X)-np.mean(X))*(np.array(Y)-np.mean(Y))) / (n*np.std(X)*np.std(Y))
print(round(pearson_coef, 3))
```

__Solution (Python)__

```python
n = int(input().strip())

X = list(map(float, input().rstrip().split()))

Y = list(map(float, input().rstrip().split()))

print(n, X, Y)
```

```python
def get_rank(x_list):
    rank = {x: i+1 for i, x in enumerate(sorted(set(x_list)))}
    return [rank[key] for key in x_list]
```

```python
X_rank = get_rank(X)
Y_rank = get_rank(Y)
sum_d_sq = sum([(x-y)**2 for x, y in zip(X_rank, Y_rank)])
r_xy = 1 - 6*sum_d_sq/(n*(n**2-1))
print(round(r_xy, 3))
```

```python
data = []

for _ in range(5):
    data.append(tuple(map(float, input().rstrip().split())))
```

```python
X = [da[0] for da in data]
Y = [da[1] for da in data]
print(X, Y)
print(data)
```

__Pure Python__

```python
n = len(data)
sum_X = sum(X)
mean_X = sum_X/n
sum_X_sq = sum([x**2 for x in X])
sum_Y = sum(Y)
mean_Y = sum_Y/n
sum_XY = sum([x*y for x, y in zip(X, Y)])
```

```python
lm_coef = (n*sum_XY-sum_X*sum_Y) / (n*sum_X_sq-sum_X**2)
lm_intercept = mean_Y - lm_coef*mean_X
```

```python
x = 80
pred = lm_coef*x + lm_intercept
print(round(pred, 3))
```

__NumPy and sklearn__

```python
from sklearn import linear_model
import numpy as np

X = np.array(X).reshape(-1, 1)

lm = linear_model.LinearRegression()
lm.fit(X, Y)
pred = lm.predict(np.array(80).reshape(-1,1))
print(lm.coef_[0], lm.intercept_)
print(round(pred[0], 3))
```

# Sampling $\pi$ 

```python
import numpy as np
from random import random
import math
from statistics import stdev

PI = math.pi
num_points = 1000
num_trials = 1000
radius = 1
sample_pi = []

for _ in range(num_trials):
    sample = [(2*random(), 2*random()) for _ in range(num_points)]
    inlier = [True for x, y in sample if math.sqrt((x-1)**2 + (y-1)**2)<radius]
    sample_pi.append(4*sum(inlier)/num_points)
mean_pi = sum(sample_pi)/num_trials
print(mean_pi)
```

```python
mu, sigma = PI, 0.05
s = np.random.normal(mu, sigma, num_trials)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1, 1)
# ax.plot(s)
ax.hist(sample_pi)
```
